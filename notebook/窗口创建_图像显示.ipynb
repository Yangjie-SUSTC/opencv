## 图像显示

```python
cv2.imshow('window title',img)
cv2.waitKey(0)# 在特定的几毫秒内等待按键，并返回该键的ASCII码，没有按键，返回-1，设为0就一直等待
#按任意键返回
#CV.distroyWindow(窗口名)删除特定窗口
cv2.destroyAllWindows()
```

```python
cv2.imshow('miao~~~',img)
k=cv2.waitKey(0)
if k==27&0xFF:    # wait for esc, 0xFF for 64bit system
    cv2.destroyAllWindows()
elif k==ord('s'):
    cv2.imwrite('../images/keysave.jpeg',img)
    cv2.destroyAllWindows()
        
```

#### matplotlib
注意：   
彩色图像使用 OpenCV 加载时是 BGR 模式。但是 Matplotib 是 RGB
模式。所以彩色图像如果已经被 OpenCV 读取，那它将不会被 Matplotib 正
确显示。

```python
from matplotlib import pyplot as plt
#%matplotlib show in line
plt.imshow(img,cmap='gray',interpolation='bicubic')
plt.xticks([]),plt.yticks([])# hide XY axis
plt.show()
```

## 创建窗口
    在这种情况下，可以调整窗口大小再显示图像

```python
cv2.namedWindow('image',cv2.WINDOW_NORMAL)# 默认值是cv2.WINDOW_AUTOSIZE
cv2.imshow('image',img) # 这里的title需要与之前创建的窗口名字一致，不然会再新建一个窗口
cv2.waitKey(0)
cv2.destroyAllWindows()
```
